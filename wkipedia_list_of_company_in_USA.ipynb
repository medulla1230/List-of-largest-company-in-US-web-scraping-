```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

##page.status_code

##len(page.text)

page_contents = page.text

##page_contents[:1000]

soup = BeautifulSoup(page_contents, 'html')
```

```python
type(soup)
```

```python
soup.find('table')
```

```python
table_tag = soup.find_all('table')[1]
```

```python
print(table_tag)
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
##table_desc = 'wikitable sortable'

##table_tag = soup.find('table', {'class' : table_desc})

##table_tag
```

```python
world_table = table_tag.find_all('th')
```

```python
world_table
```

```python
world_table_header = [title.text.strip() for title in world_table]
```

```python
print(world_table_header)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_header)

df
```

```python
column_data = table_tag.find_all('tr')
```

```python
for row in column_data[1:]:

    row_data = row.find_all('td')
    
    individual_row_data = [data.text.strip() for data in row_data]

    ##print(individual_row_data)

    lenght = len(df)
    df.loc[lenght] = individual_row_data
    
```
